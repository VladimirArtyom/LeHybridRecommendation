```python
import pandas as pd
import numpy as np
from textblob import TextBlob

df = pd.read_csv("reviews_filtered.csv")
df.head()

df["sentiment"] = df["Review"].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
def getRecommendationMostPopularSimple(df: pd.DataFrame):
    copy_df = df.copy()
    copy_df = copy_df.sort_values("Rating", ascending=False)
    return copy_df

def getRecommendationMostPopularPolarity(df: pd.DataFrame, rating_weight: float,
                                         sentiment_weight: float):
    copy_df = df.copy()
    copy_df["wr"] = copy_df["Rating"] * rating_weight + copy_df["sentiment"] * sentiment_weight
    copy_df = copy_df.sort_values("wr", ascending=False)
    return copy_df


getRecommendationMostPopularPolarity(df, 0.6, 0.4)
```
