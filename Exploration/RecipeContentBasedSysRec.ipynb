```python
# Potential Contents :  Description, RecipeInstructions, AuthorName, Keywords, RecipeCategory, TotalTime,
column_of_interest = ["Name","Description", "RecipeInstructions", "AuthorName", "Keywords", "RecipeCategory", "TotalTime"]
```

```python
def cleanStr(x : str):
    x = x.strip()
    x = x.replace("'","")
    x = x.replace('"',"")
    x.lower()
    return x

def cStrToList(x: str):
    x = x[2:-1]
    x = x.replace('"', "" )
    return [cleanStr(s) for s in x.split(',')]

def listTostr(t):
    mainStr: str = ""
    for indx in range(len(t)):
        mainStr += " " + t[indx].lower()
    return mainStr
def lower_case(t: str):
    return t.lower()

def mergeAuthorName(x: str):
    x = x.lower()
    return x.strip().replace(" ", "-").replace(".","")
```

```python
df_copy: pd.DataFrame = df.copy()
df_copy.fillna("[]", inplace=True)
df_copy = df_copy[column_of_interest]
df_copy["Name"] = df_copy["Name"].apply(lower_case)
df_copy["Description"] = df_copy["Description"].apply(lower_case)
df_copy["RecipeCategory"] = df_copy["RecipeCategory"].apply(lower_case)
df_copy["TotalTime"] = df_copy["TotalTime"].apply(lower_case)
df_copy["AuthorName"] = df_copy["AuthorName"].apply(mergeAuthorName)
df_copy["RecipeInstructions"] = df_copy["RecipeInstructions"].apply(cStrToList).apply(listTostr)
df_copy["Keywords"] = df_copy["Keywords"].apply(cStrToList).apply(listTostr)
```

```python
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load GloVe vectors (change path to your GloVe file)
glove_file = "GoogleNews-vectors-negative300.bin"
word_vectors = KeyedVectors.load_word2vec_format(glove_file, binary=True)
```

```python


def sentence_embedding(sentence):
    # Tokenize the sentence and get GloVe vectors for each word
    tokens = sentence.lower().split()
    vectors = [word_vectors[word] for word in tokens if word in word_vectors]

    # If no words found in vocabulary, return None
    if not vectors:
        return None

    # Calculate the mean of word vectors to get sentence embedding
    return np.mean(vectors, axis=0)

def sentence_similarity(sentence1, sentence2):
    # Get embeddings for both sentences
    embedding1 = sentence_embedding(sentence1)
    embedding2 = sentence_embedding(sentence2)

    # If any of the sentences couldn't be embedded, return None
    if embedding1 is None or embedding2 is None:
        return None

    # Calculate cosine similarity between sentence embeddings
    sim = cosine_similarity([embedding1], [embedding2])[0][0]
    return sim

# Example usage
sentence1 = "I love dogs"
sentence2 = "I like cats"
similarity = sentence_similarity(sentence1, sentence2)
print("Similarity between the sentences:", similarity)
```
