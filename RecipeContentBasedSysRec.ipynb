```python
# Potential Contents :  Description, RecipeInstructions, AuthorName, Keywords, RecipeCategory, TotalTime,
column_of_interest = ["Name","Description", "RecipeInstructions", "AuthorName", "Keywords", "RecipeCategory", "TotalTime"]
```

```python
def cleanStr(x : str):
    x = x.strip()
    x = x.replace("'","")
    x = x.replace('"',"")
    x.lower()
    return x

def cStrToList(x: str):
    x = x[2:-1]
    x = x.replace('"', "" )
    return [cleanStr(s) for s in x.split(',')]

def listTostr(t):
    mainStr: str = ""
    for indx in range(len(t)):
        mainStr += " " + t[indx].lower()
    return mainStr
def lower_case(t: str):
    return t.lower()

def mergeAuthorName(x: str):
    x = x.lower()
    return x.strip().replace(" ", "-").replace(".","")
```

```python
df_copy: pd.DataFrame = df.copy()
df_copy.fillna("[]", inplace=True)
df_copy = df_copy[column_of_interest]
df_copy["Name"] = df_copy["Name"].apply(lower_case)
df_copy["Description"] = df_copy["Description"].apply(lower_case)
df_copy["RecipeCategory"] = df_copy["RecipeCategory"].apply(lower_case)
df_copy["TotalTime"] = df_copy["TotalTime"].apply(lower_case)
df_copy["AuthorName"] = df_copy["AuthorName"].apply(mergeAuthorName)
df_copy["RecipeInstructions"] = df_copy["RecipeInstructions"].apply(cStrToList).apply(listTostr)
df_copy["Keywords"] = df_copy["Keywords"].apply(cStrToList).apply(listTostr)
```

```python
df_copy.head()
```

```python
df_copy["contents"] = df_copy["Description"] + df_copy["RecipeInstructions"] + df_copy["AuthorName"] + df_copy["Keywords"] +  df_copy["RecipeCategory"] +df_copy["TotalTime"]
```

```python
df_copy["contents"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import linear_kernel
tfIdf: TfidfVectorizer = TfidfVectorizer(analyzer="word", ngram_range=(1,2), min_df=0.0, stop_words="english")
ctVect: CountVectorizer = CountVectorizer()

tfidf_mat = tfIdf.fit_transform(df_copy["contents"])
print(tfidf_mat.shape)

cosine_sim = linear_kernel(tfidf_mat, tfidf_mat) # Calculate the similarityh for each recipe
print(cosine_sim[0])
```

```python
newPairCopy = df_copy.copy().reset_index()
newPairCopy = newPairCopy.set_index("Name")

indices = pd.Series( newPairCopy["index"], index=newPairCopy.index)
indices
```

```python
recipes: pd.Series = pd.Series(newPairCopy.index)
def get_recommendation(title: str, indices = indices, recipes_series:pd.Series = recipes):
    title = title.lower()
    indx: str = indices[title]
    similarity = list(enumerate(cosine_sim[indx]))
    similarity = sorted(similarity, key=lambda x: np.average(x[1]), reverse=True)
    print(similarity)
    similarity = similarity[1:10]
    recipe_indices : str = [i[0] for i in similarity]
    return recipes.iloc[recipe_indices]
get_recommendation("napoleons")
```
